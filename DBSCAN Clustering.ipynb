```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
X = dataset.iloc[:, [3,4]].values
X
```

```python
# Number of clusters (excluding noise points labeled as -1)
n_clusters = len(set(y_DB)) - (1 if -1 in y_DB else 0)

print("Estimated number of clusters:", n_clusters)
```

```python
plt.figure(figsize=(8,6))

# Unique cluster labels
unique_labels = np.unique(y_DB)

for i in unique_labels:
    
    if i == -1:
        # Noise points
        plt.scatter(X[y_DB == i, 0],
                    X[y_DB == i, 1],
                    c='black',
                    marker='x',
                    s=80,
                    label='Noise')
    else:
        # Cluster points
        plt.scatter(X[y_DB == i, 0],
                    X[y_DB == i, 1],
                    s=70,
                    label=f'Cluster {i+1}')

plt.title(f"DBSCAN Clustering\nEstimated number of clusters: {n_clusters}")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
DB = DBSCAN(eps=3, min_samples=2).fit(X)
y_DB=DB.fit_predict(X)
```

```python
supervised = pd.DataFrame(dataset)
```

```python
supervised['cluster_group']=y_DB
supervised
```
