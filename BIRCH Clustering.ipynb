```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
X = dataset.iloc[:, [3,4]].values
X
```

```python
from sklearn.cluster import Birch
brc=Birch(n_clusters=None)
y_brc=brc.fit_predict(X)
```
